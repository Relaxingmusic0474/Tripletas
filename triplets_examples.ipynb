# Tripletas

```python
import pandas as pd
from triplet import Triplet
```

```python
T1 = Triplet(2, 4, 5)
T2 = Triplet(0, 5, 10)
```

```python
addition = T1+T2
print(addition)
```

```python
substraction = T1-T2
print(substraction)
```

```python
product = T1*T2
print(product)
```
